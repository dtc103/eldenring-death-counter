```python
import multiprocessing
import time

class MyProcessClass:
    def __init__(self):
        self.event = multiprocessing.Event()
        self.shared_value = multiprocessing.Value('i', 0)

    def run_process(self):
        """Method to be run in a separate process."""
        print("Process started")
        while not self.event.is_set():
            with self.shared_value.get_lock():
                self.shared_value.value += 1
            print("Value incremented to:", self.shared_value.value)
        print("Event set, process will stop")

    def start_process(self):
        """Starts the multiprocessing."""
        self.process = multiprocessing.Process(target=self.run_process)
        self.process.start()

    def stop_process(self):
        """Stops the multiprocessing."""
        self.event.set()
        self.process.join()
        print("Process stopped")

process_class = MyProcessClass()
process_class.start_process()

# Let the child process run for a few seconds
time.sleep(5)

# Stop the process
process_class.stop_process()
print("Main process ended")
```
